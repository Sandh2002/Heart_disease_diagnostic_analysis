```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# importing the dependencies
# Data Collection And processing
#Loading the data into the Pandas Data Frame
heart_data =pd.read_csv('/content/heart_disease_data.csv')
#number of colum and rows in the dataset
heart_data.shape
#
#getting some information about the data
heart_data.isnull().sum()
heart_data.describe()

#
#
heart_data['target'].value_counts()
# 1--> here 1 reprent defective heart
# 0--> represent healthy heart
# splling the features(other information rather than the target)and target
X= heart_data.drop(columns='target',axis=1)
Y= heart_data['target']
print(X)
print(Y)
# splitting tha data into training data and Test data
X_train,X_test, Y_train, Y_test = train_test_split(X ,Y, test_size =0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
# Model Training
# logitistics Regression
model =LogisticRegression()
model.fit(X_train, Y_train)
# model evaluation
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```
